```python
import numpy as np
import pandas as pd
import sklearn.svm as svm
from sklearn.metrics import accuracy_score
from plotSVMBoundaries import plotSVMBoundaries
```

```python
#main function
def support_vector_machine(train_x, train_y,test_x,test_y,c,k='linear',G='scale'):
    clf=svm.SVC(C=c,kernel =k,gamma =G)
    clf.fit(train_x,train_y)
    y_test_pred=clf.predict(test_x)
    y_train_pred=clf.predict(train_x)
    test_acc=accuracy_score(test_y,y_test_pred)
    train_acc=accuracy_score(train_y,y_train_pred)
    plotSVMBoundaries(train_x,train_y,clf,support_vectors=clf.support_vectors_)
    plotSVMBoundaries(test_x,test_y,clf)
    print("Test Accuracy :",test_acc*100)
    print("Train Accuracy :",train_acc*100)
    return clf
```

```python
#Loading the datasets
df1 = pd.read_csv('/content/drive/MyDrive/EE559/Code/dataset/HW1_datasets/dataset1_train.csv',header=None)
df2 = pd.read_csv('/content/drive/MyDrive/EE559/Code/dataset/HW1_datasets/dataset1_test.csv',header=None)
```

```python
#converting the data to numpy
array1=df1.to_numpy()
array2=df2.to_numpy()
#slicing the data
X_train=array1[:,:-1]
Y_train=array1[:,-1]
X_test=array2[:,:-1]
Y_test=array2[:,-1]
```

C=0.01

```python
clf = support_vector_machine(X_train, Y_train,X_test,Y_test,0.01,k='linear',G='scale')
print("Weight :",clf.coef_)
print("Bais :",clf.intercept_)
```
